Top 5 Data Visualization Libraries in Python:

Matplotlib:

Features:Matplotlib is a versatile library for creating static, animated, and interactive visualizations in Python. It offers fine-grained control over every aspect of a plot, allowing users to create highly customized charts and graphs. Matplotlib supports various plot types, including line plots, scatter plots, bar plots, histograms, and more.
Applications: Matplotlib is widely used for data exploration, scientific plotting, statistical analysis, and publication-quality graphics.


Seaborn:

Features: Seaborn is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics. It simplifies the process of creating complex visualizations by providing built-in themes, color palettes, and functions for common statistical plots, such as distribution plots, violin plots, and regression plots.
Applications: Seaborn is commonly used for data visualization in statistical analysis, exploratory data analysis (EDA), and data-driven storytelling.


Plotly:

Features: Plotly is a versatile library for creating interactive visualizations in Python. It supports a wide range of chart types, including scatter plots, line plots, bar plots, histograms, 3D plots, and more. Plotly's interactive plots can be embedded in web applications, Jupyter notebooks, and standalone HTML documents.
Applications: Plotly is used for creating interactive dashboards, web-based applications, data exploration, and sharing visualizations online.


Bokeh:

Features: Bokeh is a Python library for creating interactive visualizations for the web. It targets modern web browsers and supports streaming, real-time data, and large datasets. Bokeh provides a high-level plotting interface similar to Matplotlib, as well as a lower-level interface for creating custom interactive visualizations using JavaScript.
Applications: Bokeh is commonly used for creating interactive dashboards, web-based applications, and data visualization in data science projects.


Altair:

Features: Altair is a declarative statistical visualization library for Python. It is based on the Vega and Vega-Lite visualization grammars, which allow for concise and declarative specification of visualizations. Altair's API is simple and intuitive, making it easy to create complex visualizations with minimal code.
Applications: Altair is used for exploratory data analysis, statistical visualization, and data visualization in research and academic settings.




Top 5 Data Visualization Libraries in R:

ggplot2:

Features: ggplot2 is an implementation of the grammar of graphics in R. It provides a powerful and flexible system for creating complex visualizations by combining simple building blocks. ggplot2 supports a wide range of plot types, including scatter plots, line plots, bar plots, histograms, and more.
Applications: ggplot2 is widely used for data visualization in data analysis, statistical modeling, and data-driven storytelling.


Plotly:

Features: Plotly is a versatile visualization library that supports both Python and R. It offers a wide range of chart types and provides an API for creating interactive visualizations with rich interactivity and customization options. Plotly's interactive plots can be embedded in web applications, R Markdown documents, and Shiny applications.
Applications: Plotly is used for creating interactive dashboards, web-based applications, and data visualization in data science projects.


Lattice:

Features: Lattice is a plotting system in R for creating multi-panel plots and conditioned plots. It provides a high-level interface for creating trellis plots, which are useful for visualizing relationships in multi-dimensional data. Lattice supports various plot types, including scatter plots, line plots, bar plots, and histograms.
Applications: Lattice is commonly used for exploratory data analysis, visualizing multivariate data, and data visualization in research and academic settings.


ggvis:

Features: ggvis is an interactive visualization package for R based on the grammar of graphics. It provides a flexible and expressive syntax for creating interactive plots with rich interactivity and animation. ggvis supports linked brushing, interactive tooltips, and interactive filtering.
Applications: ggvis is used for creating interactive visualizations in exploratory data analysis, data visualization in web applications, and interactive data storytelling.


rCharts:

Features: rCharts is an R package for creating interactive visualizations using JavaScript libraries such as Highcharts, NVD3, and Leaflet. It provides an R interface for generating interactive plots with rich interactivity and customization options. rCharts supports a wide range of chart types, including line charts, bar charts, scatter plots, and maps.
Applications: rCharts is used for creating interactive dashboards, web-based applications, and data visualization in data science projects.

```r
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

def add_percentage_labels(ax):
    total = len(data)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')

sns.countplot(ax=axes[0, 0], x='Fuel_Type', data=data, color='blue',
              order=data['Fuel_Type'].value_counts().index)
add_percentage_labels(axes[0, 0])

sns.countplot(ax=axes[0, 1], x='Transmission', data=data, color='blue',
              order=data['Transmission'].value_counts().index)
add_percentage_labels(axes[0, 1])

sns.countplot(ax=axes[1, 0], x='Owner_Type', data=data, color='blue',
              order=data['Owner_Type'].value_counts().index)
add_percentage_labels(axes[1, 0])

sns.countplot(ax=axes[1, 1], x='Location', data=data, color='blue',
              order=data['Location'].value_counts().index)
add_percentage_labels(axes[1, 1])


axes[1][1].tick_params(labelrotation=45)
axes[2][0].tick_params(labelrotation=90)
axes[2][1].tick_params(labelrotation=90)

plt.show()
```

```r
install.packages("gridExtra")
install.packages("ggplot2")

library(gridExtra)
library(ggplot2)

# Assuming 'data' is your dataset

# Define a function to add percentage labels to the plot
add_percentage_labels <- function(p) {
  total <- length(data[[p]])
  p <- ggplot(data, aes_string(x = p)) +
    geom_bar(fill = "blue") +
    geom_text(stat='count', aes(label = scales::percent(..count../total)),
              position = position_stack(vjust = 0.5)) +
    labs(title = paste("Bar plot for", p),
         x = p,
         y = "Count") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

  return(p)
}

# Create a list of plots
plots <- list()

# Plot for Fuel_Type
plots[[1]] <- add_percentage_labels("Fuel_Type")

# Plot for Transmission
plots[[2]] <- add_percentage_labels("Transmission")

# Plot for Owner_Type
plots[[3]] <- add_percentage_labels("Owner_Type")

# Plot for Location
plots[[4]] <- add_percentage_labels("Location")

# Arrange and display plots
gridExtra::grid.arrange(grobs = plots, ncol = 2)
```
