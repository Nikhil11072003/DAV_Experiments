```r

# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import string
import requests
from bs4 import BeautifulSoup

```

```r
import nltk
nltk.download('punkt')
```

```r
# Tokenization
text = "This is a sample text. Tokenization is the first step in text analytics."
sentences = sent_tokenize(text)
words = word_tokenize(text)
```

```r
import nltk
nltk.download('stopwords')
```

```r
# Frequency Distribution
fdist = nltk.FreqDist(words)

# Removing Stopwords & Punctuations
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words and word not in string.punctuation]
```

```r

import nltk
nltk.download('wordnet')
```

```r
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```r

import nltk
nltk.download('averaged_perceptron_tagger')
```

```r
pos_tags = nltk.pos_tag(filtered_words)
```

```r
import requests
from bs4 import BeautifulSoup

# Define the URL you want to scrape
url = 'https://www.geeksforgeeks.org/python-programming-language/'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all 'a' tags (links) in the HTML
    links = soup.find_all('a')

    # Extract and print the URLs
    for link in links:
        href = link.get('href')
        if href is not None:
            print(href)
else:
    print('Failed to retrieve the webpage:', response.status_code)
```

```r
import requests
from bs4 import BeautifulSoup

# Define the URL you want to scrape
url = 'https://www.geeksforgeeks.org/python-programming-language/'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all heading elements (h1, h2, h3, etc.) in the HTML
    headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

    # Extract and print the text content of the headings
    for heading in headings:
        print(heading.text.strip())
else:
    print('Failed to retrieve the webpage:', response.status_code)
```
